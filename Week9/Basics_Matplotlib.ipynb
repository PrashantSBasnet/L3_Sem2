**Basics of Data Visualization**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Line Plot**


```python
views = [123, 234, 456, 890, 454, 665]
plt.plot(views)
plt.show()
```

```python
views = [123, 234, 456, 890, 454, 665, 777]
days = [1,2,3,4,5,6,7] #alternatively, days=range(1,8)    #this is for X-Axis

#plt.plot(X,Y)
plt.plot(days, views)  #plt.plot(X-Axis, Y-Axis)
plt.xlabel('Day in Number')
plt.ylabel('Number of sales')
plt.show()
```

```python
views = [123, 234, 456, 890, 454, 665, 777]
days = [1,2,3,4,5,6,7] #alternatively, days=range(1,8)    #this is for X-Axis

#plt.plot(X,Y)
plt.plot(days, views, label='Sales')  #plt.plot(X-Axis, Y-Axis)
plt.xlabel('Day in Number')
plt.ylabel('Number of sales')
plt.legend() #plt.legend(loc='upper right')
plt.title('Line Plot representing weekly sales')
plt.show()
```

**Bar Plot**

```python
# Days of the week
days = ['Sun', 'Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat']

# Corresponding x-axis values
x = range(len(days))

# Y-axis values
y = [10, 5, 8, 4, 2,4,5]

# Plotting the bar chart with day labels
plt.bar(x, y, tick_label=days, color='g')

# Labeling the x-axis
plt.xlabel('Days')

# Labeling the y-axis
plt.ylabel('No. of sales')

# Title of the plot
plt.title('Bar Chart representing no. of sales per by day')

# Displaying the plot
plt.show()
```

**Scatter Plot**

```python
# Days of the week
days = ['Sun', 'Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat']

# Corresponding x-axis values
x = range(len(days))

# Y-axis values
y = [10, 5, 8, 4, 2, 4, 2]

# Labeling the x-axis
plt.xlabel('Days')

# Labeling the y-axis
plt.ylabel('No. of sales')

# Function to plot scatter
plt.scatter(x, y)

# function to show the plot
plt.show()
```

**Histogram**

```python
# Generate random data for the histogram
data = np.random.randn(1000)

# Plotting a basic histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

# Display the plot
plt.show()
```
