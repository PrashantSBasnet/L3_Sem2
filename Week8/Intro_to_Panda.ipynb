**Panda Series**
*   One-dimensional array holding data of any type.
*   A fundamental data structure in the pandas library, which is widely used for data manipulation and analysis in Python

Key features:

1.   **Indexing**: Each element in a Series is associated with an index. By default, the index is a sequence of integers starting from 0, but you can also specify custom index labels.
2.   **Homogeneity**: All elements in a Series must be of the same data type. Pandas automatically infers the data type of the Series based on its contents.
3.   **Size Mutability**: You can modify the size (length) of a Series after it has been created by adding or removing elements.
4.   **Vectorized Operations**: Pandas Series supports vectorized operations, allowing you to perform element-wise operations on the entire Series without using explicit loops.
5.   **Alignment and Broadcasting**: When performing operations between two Series objects, pandas aligns them based on their index labels, ensuring that corresponding elements are matched correctly. This enables broadcasting of operations across Series with different lengths and index labels.

A Pandas **DataFrame**
*   DataFrame is a two-dimensional labeled data structure in the pandas library for Python. It is similar to a spreadsheet or SQL table, where data is organized into rows and columns.
*   Each column in a DataFrame can have a different data type (e.g., integer, float, string) and is labeled with a column name, while each row is labeled with an index.


Characteristics:
1. **Indexing**: Each row and column in a DataFrame is associated with an index. The row index represents the label for each row, and the column index represents the label for each column. By default, row indices start from 0 and column indices start from 0, but you can also specify custom labels.
2. **Heterogeneity**: Each column in a DataFrame can have a different data type. Pandas automatically infers the data types of columns based on their contents.
3. **Size Mutability**: You can modify the size (number of rows and columns) of a DataFrame after it has been created by adding or removing rows and columns.
4. **Alignment and Broadcasting**: When performing operations between two DataFrames or between a DataFrame and a Series, pandas aligns them based on their indices, ensuring that corresponding elements are matched correctly. This enables broadcasting of operations across different shapes of DataFrames and Series.
5. **Powerful Data Manipulation**: Pandas provides a wide range of functions and methods for data manipulation, including filtering, selecting, sorting, grouping, aggregating, merging, and joining.

**Operations on Rows**

```python
#Retrieving rows

# Define a dictionary containing employee data
data = {'Name':['Nirnaya', 'Nikita', 'Manish', 'Kanchana'],
        'Age':[18, 19, 20, 19],
        'Address':['Ktm', 'Thapathali', 'Baneshwor', 'Tinkune'],
        'Domain':['Python', 'Java', '.Net', 'Angular']}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)

# Set the index of the DataFrame to the 'Name' column
df.set_index('Name', inplace=True)

# retrieving row by loc method
first = df.loc["Nirnaya"]
x = df.loc["Manish"]

print (first, "\n\n\n",x)


```

**Adding Rows**

*   We can add a single row using DataFrame.loc
*   We can add the row at the last in our dataframe.
*   We can get the number of rows using len(DataFrame.index) for determining the position at which we need to add the new row.




```python
dict = {'Name':['Pratyush', 'Manish', 'Evanjellyna', 'Enyous'],
        'Sports':['Football', 'BasketBall', 'TableTennis', 'VolleyBall'],
        'LunchType':['Veg', 'NonVeg', 'Veg', 'NonVeg']
       }

df = pd.DataFrame(dict)

display(df)
```

```python
# Access the last row using loc[]
last_row = df.loc[len(df.index) - 1]

print(last_row)
```

```python
df.loc[len(df.index)] = ['Dilasha', 'Cricket', 'NonVeg']

display(df)
```
